```python
import pandas as pd
```

```python
df = pd.read_csv("/Workspace/Users/saeedsoliman901@gmail.com/supply_chain_project/DataCoSupplyChainDataset.csv", encoding='ISO-8859-1')
```

```python
#clean columns names
df.columns = [col.replace(' ', '_').replace('(', '').replace(')', '').replace('-', '_') for col in df.columns]
```

```python
#clean data in columns
text_cols = ['Customer_City', 'Customer_Country', 'Order_City', 'Order_Country', 'Product_Name', 'Category_Name']
for col in text_cols:
    df[col] = df[col].str.strip().str.title()
```

```python

order_date_col = [c for c in df.columns if 'order_date' in c.lower()][0]
shipping_date_col = [c for c in df.columns if 'shipping_date' in c.lower()][0]
```

```python
#transform order_date and shipping_date to datetime
df['order_date_dt'] = pd.to_datetime(df[order_date_col])
df['shipping_date_dt'] = pd.to_datetime(df[shipping_date_col])
```

```python
#measure shipping gap between scheduled and real
df['Shipping_Gap'] = df['Days_for_shipping_real'] - df['Days_for_shipment_scheduled']
```

```python
#drop columns
cols_to_drop = [
    'Product_Card_Id', 'Product_Image', 'Customer_Email', 'Customer_Password', 
    'Customer_Fname', 'Customer_Lname', 'Customer_Street', 'Customer_Zipcode',
    'Product_Description', 'Order_Zipcode', 'order_date_Date_Orders', 'shipping_date_Date_Orders'
]
```

```python
# to skip errors
df.drop(columns=cols_to_drop, inplace=True, errors='ignore')
```
